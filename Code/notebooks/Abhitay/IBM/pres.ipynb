```python
import pandas as pd

def pretty_describe(df):
    """
    Returns a more readable summary of the dataset.
    - Rounds values for better readability.
    - Formats large numbers with commas.
    - Converts scientific notation to standard notation.
    """

    summary = df.describe().transpose()

    # ✅ Formatting large numbers with commas
    summary = summary.applymap(lambda x: f"{x:,.2f}" if isinstance(x, (int, float)) else x)

    # ✅ Rename the index column to "Feature"
    summary.index.name = "Feature"

    # ✅ Rename the rows for clarity
    summary = summary.rename(index={
        "count": "Total Records",
        "mean": "Mean",
        "std": "Standard Deviation",
        "min": "Minimum",
        "25%": "25th Percentile",
        "50%": "Median (50th Percentile)",
        "75%": "75th Percentile",
        "max": "Maximum"
    })

    return summary

# ✅ Example Usage
pretty_summary = pretty_describe(df)

# ✅ Display in Jupyter Notebook
from IPython.display import display
display(pretty_summary)  # Works in Jupyter Notebook


# ✅ Print as Table (For Terminal Use)
print(pretty_summary.to_markdown())  # Requires `pip install tabulate`
```
